```python
# import resources
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import cv2

%matplotlib inline
```

```python
# read image 

## first image
stop1 = mpimg.imread('media/stop_sign.jpg')
print('First image shape: {}'.format(stop1.shape))

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.title('First Image')
plt.imshow(stop1)

## second image
stop2 = mpimg.imread('media/stop_sign2.jpg')
print('Second Image shape: {}'.format(stop2.shape))
plt.subplot(122)
plt.title("Second Image")
plt.imshow(stop2)
```

```python
# crop second image to resemble first image

## make a copy of the image to manipulate
image_crop = np.copy(stop2)

## define how many pixels to slice off the sides of the original image
row_crop = 90
col_crop = 250

# using image slicing, subtract the row_crop from the top/bottom and col_crop from left/right
image_crop = stop2[row_crop:-row_crop, col_crop:-col_crop, :]

plt.figure(figsize=(15,5))
plt.subplot(121)
plt.title('Original image')
plt.imshow(stop2)

plt.subplot(122)
plt.title('Cropped image')
plt.imshow(image_crop)
```

```python
# resize cropped image to be the same as first

## use opencv resize function
standardized_im = cv2.resize(image_crop, (stop1.shape[1], stop1.shape[0]))     # shape = [x, y, c]

print('Shape of Standardized image: {}'.format(standardized_im.shape))

# plot both the images side by side
plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.title('First image')
plt.imshow(stop1)

plt.subplot(122)
plt.title('Standardized Second image')
plt.imshow(standardized_im)
```

```python
# compare both images

## sum all the red channel values and compare
red_sum1 = np.sum(stop1[:, :, 0])
red_sum2 = np.sum(standardized_im[:, :, 0])

print('Sum of all red pixel values in the first stop sign image: ', red_sum1)
print('Sum of all red pixel values in the second standarized image: ', red_sum2)

red_sum_original = np.sum(stop2[:, :, 0])
print('\nSum of all red pixel values in the second non-standardized image: ', red_sum_original)
```
